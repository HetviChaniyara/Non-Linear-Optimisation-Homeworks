```python
import numpy as np
import matplotlib.pyplot as plt
```

Part a) Plots in 2D


```python
n = 250
x = np.linspace(0,2,n)

# functions
f = np.exp(-3 * x) * np.sin(20 * x)
g = np.where(x<=1, x**(1/4), 1)

# plots
plt.plot(x, f, label=r'f(x)', color='blue')
plt.plot(x, g, label=r'g(x)', color='red')

plt.title('Plot of f(x) & g(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

Part b) Plots in 3D

```python
import plotly.graph_objects as go
J = np.linspace(-2 * np.pi, 2 * np.pi, n)
X,Y = np.meshgrid(J,J)

# function
h = np.sin(X + Y) * np.sin(X - Y)

fig = go.Figure(data=[go.Surface(z=h, x=X, y=Y)])

fig.update_layout(title='3D Surface Plot', autosize=False)
fig.show()
```

# Part c)

```python
x = np.linspace(-2, 2, n)
y = np.linspace(-1, 3, n)
X, Y = np.meshgrid(x, y)

W = 100 * (Y - X**2)**2 + (1 - X)**2

levels = np.logspace(np.log10(W.min()), np.log10(W.max()), 50)

plt.figure(figsize=(10, 6))
cp = plt.contour(X, Y, W, levels=levels, cmap='viridis')
plt.colorbar(cp, label='w(x, y) value (log scale)')

plt.title('Contour Plot')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
